```python
# Datos

k = 14.5  #[m/d]  Conductividad hidraulica
R = 0.007 #[m/d]  Recarga
h1 = 17.6  #[m]    altura (medida desde la base impermeable)
h2 = 15.3  #[m]    altura (medida desde la base impermeable)
L = 550    #[m]    Distancia entre los 2 cuerpos de agua (h1-h2)

x1 = 0
x2 = 525
x3 = 550

# Caudal unitario
q1 = ((h1**2 - h2**2) * k) / (2 * L) - R * (L / 2 - x1)
q2 = ((h1**2 - h2**2) * k) / (2 * L) - R * (L / 2 - x2)
q3 = ((h1**2 - h2**2) * k) / (2 * L) - R * (L / 2 - x3)

# Divisoria de aguas
d = (L / 2) - ((h1**2 - h2**2) * k) / (2 * L * R)

# Altura de la divisoria de aguas
h = (((h2**2 - h1**2) * d / L) + (R * d / k) * (L - d) + h1**2)**0.5

print(q1,q2,q3,d,h)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Valores iniciales y finales de x
xi = 0
xf = 550
# Valores de x en el rango especificado
x = np.linspace(xi, xf, 10)

# Parámetros de la ecuación
h1= 17.6
h2= 15.3
R = 0.007
k = 14.5
L = xf

# Calcula las alturas h para cada valor de x
h = [( (((h2**2-h1**2)*x)/L) + (((R*x)/k)*(L-x)) + h1**2 )**0.5 for x in x]

# Crea la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, h, label='Altura del nivel freático (h)')

# Puntos iniciales y finales
plt.scatter([0, 550], [17.6, 15.3], color='red', label='Valores semilla')

# Etiquetas y leyendas
plt.xlabel('Distancia (x)')
plt.ylabel('Altura del nivel freático (h)')
plt.title('Altura del Nivel Freático en función de la Distancia')
plt.legend()

# Muestra la gráfica
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Valores iniciales y finales de x
xi = 0
xf = 550
# Valores de x en el rango especificado
x = np.linspace(xi, xf, 10)

# Parámetros de la ecuación
h1= 17.6
h2= 15.3
R = 0.007
k = 14.5
L = xf

# Calcula las alturas h para cada valor de x
h = [( (((h2**2-h1**2)*x)/L) + (((R*x)/k)*(L-x)) + h1**2 )**0.5 for x in x]

# Crea la gráfica
fig, ax = plt.subplots(figsize=(10, 6))

# Puntos iniciales y finales
ax.scatter([0, 550], [17.6, 15.3], color='red', label='Valores semilla')

# Puntos de la curva
points, = ax.plot(x, h, 'o', picker=5) # 5 points tolerance

def onpick(event):
    # event.ind es a list of indices into the original MaskedArray object
    # of the points picked. Here, we find the index of the picked point.
    ind = event.ind[0]
    print(f'Punto seleccionado: x={x[ind]:.2f}, h={h[ind]:.2f}')

fig.canvas.mpl_connect('pick_event', onpick)

# Etiquetas y leyendas
ax.set_xlabel('Distancia (x)')
ax.set_ylabel('Altura del nivel freático (h)')
ax.set_title('Altura del Nivel Freático en función de la Distancia')
ax.legend()
```
